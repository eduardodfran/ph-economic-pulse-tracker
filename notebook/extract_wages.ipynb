```python
import tabula
import pandas as pd

# Path to your downloaded PDF
pdf_path = "../data/wage-data.pdf"

# Read PDF into a list of DataFrames
dfs = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)

# Concatenate all pages into one master wage table
df_wages = pd.concat(dfs)

# Cleaning: You'll need to strip strings like "WO" or "Region" to make it joinable
df_wages.to_csv("stg_min_wage.csv", index=False)
```
