# Inspect Poverty Dataset Columns and Data

This notebook reads the local Philippines poverty CSV from the repository and validates columns, dtypes, and sample values before finalizing ingestion schema.

```python
import pandas as pd
from pathlib import Path
```

```python
POVERTY_FILE = Path("../data/poverty_phl.csv")
df_raw = pd.read_csv(POVERTY_FILE)
df_raw.head()
```

```python
clean_columns = [c.lower().strip().replace(' ', '_').replace('.', '') for c in df_raw.columns]
df = df_raw.copy()
df.columns = clean_columns
print("Column count:", len(df.columns))
print(df.columns.tolist())
```
